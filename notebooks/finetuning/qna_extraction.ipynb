```python
import pandas as pd
import json

# 1. 파일 불러오기
qna_df = pd.read_excel('data/merged_QnA_final.xlsx')
meta_df = pd.read_excel('data/merged_metadata_final.xlsx')

dataset = []

# 2. Q&A 데이터 변환 (전문 지식 + 상담 말투)
for _, row in qna_df.iterrows():
    instruction = "너는 다정하고 전문적인 펫 상담사야. 보호자의 질문에 공감하고 논리적으로 답변해줘."
    question = row['질문']
    # 논리적 답변 구조: 공감 + 답변 + 마무리
    answer = (
        f"안녕하세요 보호자님! 아이 때문에 걱정이 많으셨겠어요. "
        f"{row['답변']} "
        f"답변이 도움이 되셨을까요? 더 궁금한 점은 언제든 물어봐 주세요!"
    )
    dataset.append({"instruction": instruction, "input": question, "output": answer})

# 3. 메타데이터 변환 (품종 지식 + 상담 말투)
for _, row in meta_df.iterrows():
    instruction = "너는 품종별 특성을 잘 아는 전문 상담사야. 아이의 품종에 맞는 조언을 해줘."
    # 질문 예시 생성: "보더 콜리의 특징이 뭐야?"
    question = f"{row['품종명(국문)']}의 특징과 관리법을 알려줘."
    
    # 메타데이터 기반 논리적 답변 생성
    answer = (
        f"네, {row['품종명(국문)']}에 대해 궁금하시군요! "
        f"이 친구들은 {row['일반 특징']}는 특징이 있어요. "
        f"특히 {row['건강 특징']}를 주의해야 하는데, 이를 위해 {row['수의 영양학적 메타 디스크립션']} "
        f"추천드리는 사료는 {row['좋아하는 사료']}이고, {row['건강제품']}를 챙겨주시면 더욱 좋아요."
    )
    dataset.append({"instruction": instruction, "input": question, "output": answer})

# 4. JSONL 파일로 저장
with open('train_dataset.jsonl', 'w', encoding='utf-8') as f:
    for entry in dataset:
        f.write(json.dumps(entry, ensure_ascii=False) + '\n')

print(f"변환 완료! 총 {len(dataset)}개의 데이터가 생성되었습니다.")
```
